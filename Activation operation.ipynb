```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True)

def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

def tanH(x):
    return (np.exp(x) - np.exp(-x))  / (np.exp(x) + np.exp(-x))

def binary(x):
    return 1 if x>0 else 0

def ReLu(x):
    return max(0.0 , x)

def leaky_ReLu(x):
    if x<0:
        return x*0.1
    else:
        return x

def softmax(x):
    e = np.exp(x)
    output = e / e.sum()
    return output


inputs = np.arange(-7,7 ,0.2)
output_binary = [binary(i) for i in inputs]
output_sigmoid = [sigmoid(i) for i in inputs]
output_tanh = [tanH(i) for i in inputs]
output_relu = [ReLu(i) for i in inputs]
output_leaky_relu = [leaky_ReLu(i) for i in inputs]

plots = [
    (inputs, output_binary, 'Binary', 'darkblue'),
    (inputs, output_sigmoid, 'Sigmoid', 'darkblue'),
    (inputs, output_tanh, 'tanH', 'darkblue'),
    (inputs, output_relu, 'ReLu', 'darkblue'),
    (inputs, output_leaky_relu, 'Leaky-ReLu', 'darkblue')
]

fig,axs = plt.subplots(2,3 ,figsize=(13,8))

for i , (x,y,label,color) in enumerate(plots,1):
    plt.subplot(2,3,i)
    plt.plot(x,y,c=color,lw=2,label=label)
    plt.title(label)
    plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Softmax  ouput  NN 3 Ouput   1

output = [.6 , 1.9 , 2.0]
result = softmax(output)
print('Prob:',result)
print('Sum:',sum(result))
```
